```python
file = "audio.wav";
x, sr = librosa.load(file, sr = 16000)
int(librosa.get_duration(x, sr)/60)
```

```python
max_part = 10
length = max_part * sr 
```

```python
a = x[21 * length: 22 * length]
d.Audio(a, rate = sr)
```

```python
e = sum(abs(a**2))
print(e)
```

```python
fig = plt.figure(figsize=(14, 8))
gplot = fig.add_subplot(211)
gplot.set_xlabel('Time')
gplot.set_ylabel('Amplitude')
gplot.plot(a)
```

```python
e = np.array([sum(abs(x[i:i+length]**2)) for i in range(0, len(x), length)])
```

```python
e
```

```python
plt.hist(e)
plt.show()
```

```python
data = pd.DataFrame(columns = ['energy','start','end'])
```

```python
thv = 400
row_index = 0
for i in range(len(e)):
    v = e[i]
    if(v>=thv):
        i = np.where(e == v)[0]
        data.loc[row_index, 'energy'] = v
        data.loc[row_index, 'start'] = i[0] * 10
        data.loc[row_index, 'end'] = (i[0] + 1) * 10
        row_index = row_index + 1
```

```python
data
```

```python
temp=[]
i=0
j=0
n=len(data) - 2
m=len(data) - 1
while(i<=n):
    j=i+1
    while(j<=m):
        if(data['end'][i] == data['start'][j]):
            data.loc[i,'end'] = data.loc[j,'end']
            temp.append(j)
            j=j+1
        else:
            i=j
            break
```

```python
data.drop(temp,axis=0,inplace=True)
```

```python
data
```

```python
start=np.array(data['start'])
end=np.array(data['end'])
for i in range(len(data)):
    if(i!=0):
        start_lim = start[i] - 5
    else:
        start_lim = start[i] 
    end_lim   = end[i]   
    filename=str(i+1) + ".mp4"
    ffmpeg_extract_subclip("boxing_match.mp4",start_lim,end_lim,targetname=filename)
```
