```python
from PIL import Image
import numpy as np
import os
import cv2
import keras
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Dense,Flatten,Dropout
from keras.models import model_from_json
```

```python
data = []
labels = []

inter = os.listdir("C:/Users/Yash Tandon/Desktop/python course/ML projects/frame-classification/dataset/interesting")
for inte in inter:
    img = cv2.imread("C:/Users/Yash Tandon/Desktop/python course/ML projects/frame-classification/dataset/interesting/" + inte)
    print(inte)
    print(img)
    file = Image.fromarray(img, 'RGB')
    rimage = file.resize((50, 50))
    data.append(np.array(rimage))
    labels.append(0)

ninter = os.listdir("C:/Users/Yash Tandon/Desktop/python course/ML projects/frame-classification/dataset/not-interesting")
for ni in ninter:
    img = cv2.imread("C:/Users/Yash Tandon/Desktop/python course/ML projects/frame-classification/dataset/not-interesting/" + ni)
    file = Image.fromarray(img, 'RGB')
    rimage = file.resize((50, 50))
    data.append(np.array(rimage))
    labels.append(1)
```

```python
classifications = np.array(data)
labels = np.array(labels)

np.save('classifications',classifications)
np.save('labels',labels)

s = np.arange(classifications.shape[0])
np.random.shuffle(s)
classifications = classifications[s]
labels = labels[s]

nclasses = len(np.unique(labels))
dlen = len(classifications)
```

```python
(x_train,x_test)=classifications[(int)(0.1*dlen):],classifications[:(int)(0.1*dlen)]
x_train = x_train.astype('float32')/255
x_test = x_test.astype('float32')/255
train_length=len(x_train)
test_length=len(x_test)

(y_train,y_test)=labels[(int)(0.1*dlen):],labels[:(int)(0.1*dlen)]

y_train=keras.utils.to_categorical(y_train,nclasses)
y_test=keras.utils.to_categorical(y_test,nclasses)
model=Sequential()
model.add(Conv2D(filters=16,kernel_size=2,padding="same",activation="relu",input_shape=(50,50,3)))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=32,kernel_size=2,padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=64,kernel_size=2,padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=2))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(500,activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(2,activation="softmax"))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,batch_size=50,epochs=100,verbose=1)
```

```python
score = model.evaluate(x_test, y_test, verbose=1)
print('\n', 'Test accuracy:', score[1])
```

```python
def convert_to_array(img):
    im = cv2.imread(img)
    img = Image.fromarray(im, 'RGB')
    image = img.resize((50, 50))
    return np.array(image)

def get_classification(label):
    if label==0:
        return "interesting"
    if label==1:
        return "not-interesting"
    
def predict_class(file):
    print("Predicting .................................")
    ar=convert_to_array(file)
    ar=ar/255
    label=1
    a=[]
    a.append(ar)
    a=np.array(a)
    score=model.predict(a,verbose=1)
    print(score)
    label_index=np.argmax(score)
    print(label_index)
    acc=np.max(score)
    fclass=get_classification(label_index)
    print(fclass)
    print("The predicted class for the frame is "+fclass+" with accuracy =    "+str(acc))
    
predict_class("1280px-Boxing080905_photoshop.jpg")
predict_class("focused_228979926-stock-photo-portrait-male-female-boxers-standing.jpg")
```
